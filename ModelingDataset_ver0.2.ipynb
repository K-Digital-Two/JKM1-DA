```python
import pandas as pd
```

#### 1. 선박 위치에서 가장 가까운 관측소 매칭 데이터(nearest_obs), 관측소 시간별 해양기상데이터 (weather_obs) 불러오기

```python
nearest_obs = pd.read_csv('./ModelingDataset/nearestObs.csv',sep=',', encoding = 'cp949',  index_col = 0)
weather_obs = pd.read_csv('./ModelingDataset/weather.csv', sep=',' , encoding = 'cp949',  index_col = 0)
```

#### 2. 선박 위치에서 가장 가까운 관측소 매칭 데이터(nearest_obs) 전처리

**2-1) nearest_obs 확인**

```python
nearest_obs
```

**2-3) 'insertTime'컬럼 데이터타입 바꿔주기**

- object to datetime

```python
nearest_obs['insertTime'] = pd.to_datetime(nearest_obs['insertTime'])
nearest_obs
```

#### 3. 관측소 시간별 해양기상데이터 (weather_obs) 전처리

**3-1) weather_obs 확인**

```python
weather_obs
```

**3-2) '관측시간'컬럼 데이터타입 바꿔주기**

- object to datetime

```python
weather_obs['관측시간'] = pd.to_datetime(weather_obs['관측시간'])
```

#### 4. dataFrame left join 시키기

- weather_nearest_obs = pd.merge(nearest_obs, weather_obs, ...)
- left_on = ('nearestObs','insertTime'), right_on = ('관측소명','관측시간')  # 키값
- how = 'inner'  # join 형태 정하기
- indicator = True  # 해당 컬럼의 값이 어느 데이터프레임을 반영했는지

```python
weather_nearest_obs = pd.merge(nearest_obs, weather_obs, left_on = ('nearestObs','insertTime'), right_on = ('관측소명','관측시간'), how = 'inner', indicator = True)
weather_nearest_obs
```

```python
weather_nearest_obs.info()
```

#### 5. join된 데이터프레임에서 필요없는 컬럼 drop

```python
weather_nearest_result = weather_nearest_obs.drop(['geometry','nearestObs','관측시간','위도','경도','_merge'], axis = 1)
```

```python
weather_nearest_result
```

```python
weather_nearest_result.columns = ['insertTime', 'mmsi', 'speed', 'shipLat', 'shipLon', 'obsName', 'tideLevel', 'windSpeed', 'windDirec', 'temp', 'pressure']
weather_nearest_result
```

#### 6. 입출항 데이터와 합치기 위해 MMSI 타입을 object로 통일시킨다

```python
weather_nearest_result['mmsi'] = weather_nearest_result['mmsi'].astype(str)
```
