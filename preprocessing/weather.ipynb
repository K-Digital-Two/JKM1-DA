```python
df1 = pd.read_csv('./기상)인천송도.csv', encoding = 'cp949')
df2 = pd.read_csv('./기상)태안.csv', encoding = 'cp949')
df3 = pd.read_csv('./기상)영흥도.csv', encoding = 'cp949')
df4 = pd.read_csv('./기상)대산.csv', encoding = 'cp949')
df5 = pd.read_csv('./기상)안산.csv', encoding = 'cp949')
df6 = pd.read_csv('./기상)평택.csv', encoding = 'cp949')
df7 = pd.read_csv('./기상)인천.csv', encoding = 'cp949')
df8 = pd.read_csv('./기상)덕적도.csv', encoding = 'cp949')

print("*** Merging multiple csv files into a single pandas dataframe ***")
df = pd.concat(
   [df1, df2, df3, df4, df5, df6, df7,df8], ignore_index=True)
# df.to_csv('./raw_weather.csv', encoding = 'cp949', sep = ",") 
```

### 1. 필요없는 컬럼 제거

- 유의파고
- 유의파주기
- 최대파고
- 최대파주기

```python
df = df.drop(df.iloc[:,5:11], axis=1)
df.drop(['시정(m)'], axis = 1, inplace = True)
```

### 2. 결측치 처리

```python
# 컬럼 중 "-" 값 갯수 확인 
(df== '-').sum()
```

```python
df = df.replace("-", None)
df = df.dropna(axis = 0).reset_index(drop=True)
(df== '-').sum()
```

### 3. 컬럼명 변경
- 전처리에 편리하도록

```python
# 전처리에 편리하게 바꾼다. 마지막에 다시 단위 포함시킨 컬럼명으로 변경예정
df.columns = ['관측소명', '관측시간', '위도', '경도','조위', '풍속', '풍향', '풍향_각도', '기온', '기압']
```

```python
df
```

### 4. 데이터타입 변경
- 관측시간
- 조위, 풍속, 풍향, 기온, 기압

```python
# 관측시간 타입변경 string -> datetime 
df['관측시간'] = pd.to_datetime(df['관측시간'])

# 조위, 풍속, 풍향, 기온, 기압 타입변경 string -> double
df = df.astype({ '조위' : 'float64', \
               '풍속': 'float64', \
               '풍향_각도':'float64', \
               '기온': 'double', \
               '기압': 'double'})
```

### 5. 날짜 기간 설정(22.11.30~22.12.28)
- 인천항만공사에서 추후 제공할 데이터 기간과 동일 => 11.30~12.30 데이터 제공받음.
- 31일 데이터만 삭제

```python
df_weather = df
df_weather = df_weather[df_weather["관측시간"].between('2022-11-30', '2022-12-31')]
df_weather = df_weather.iloc[:-1, :].reset_index(drop=True)
```

```python
df_weather
```

### 6. 풍향_각도 컬럼 삭제
- 모델 학습 시 '풍향'컬럼을 원핫인코딩 

```python
df_weather.drop(['풍향_각도'], axis = 1, inplace = True)
```

```python
df_weather
```

### 7. 컬럼명 원상복구

```python
df_weather.columns = ['관측소명', '관측시간', '위도', '경도', '조위(cm)', '풍속(m/s)', '풍향', '기온(℃)','기압(hPa)']
```
