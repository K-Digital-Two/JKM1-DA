### 도착지까지 거리 계산 
- rest_dist 피쳐 만들기
- from haversine import haversine, Unit

```python
from haversine import haversine, Unit
def rest_dist():
    arrival_port = (37.440515,126.601098)
    
    shipPosition = []
    for i in range(len(dt)):
        position = (dt['shipLat'][i],dt['shipLon'][i])
        shipPosition.append(position)
        
    restDist = []
    for i in range(len(shipPosition)):
        try:
            rest_dist = haversine(arrival_port, shipPosition[i], unit = 'm')
            restDist.append(rest_dist)
        except:
            "None" # 이상치 null 처리 
    dt['rest_dist'] = pd.DataFrame(restDist)
    dt.dropna(subset = ['rest_dist'], axis = 0, inplace = True)
    return dt
```

```python
rest_dist()
```

```python
from collections import Counter
shiplst = []
def select_ship():
    shipcnt = Counter(dt['mmsi'])
    model_ship = shipcnt.most_common(n=43)
    for i in range(len(model_ship)):
        shiplst.append(model_ship[i][0])
    return dt
```

```python
select_ship()
```

```python
dt = dt[dt['mmsi'].isin(shiplst)].reset_index(drop=True)
dt
```

**insertTime 날짜, 시간 분리**
- 월, 일, 시간, 요일
- 시, 분

```python
import datetime
m = []
d = []
w = []
h = []
mm = []
def time_split(time):
    for i in range(len(time)):
        m.append(time[i].month)
        d.append(time[i].day)
        h.append(time[i].hour)
        mm.append(time[i].minute)
        w.append(time[i].weekday())
    return 
```

```python
dt['insertTime']
```

```python
time_split(dt['insertTime'])
```

```python
dt['insertMonth'] = m
dt['insertDate'] = d
dt['insertHour'] = h
dt['insertMinute'] = mm
dt['insertDay'] = w
```

**arrivalTime 날짜, 시간 분리**
- 월, 일, 시간, 요일
- 시, 분

```python
time_split(dt['arrivalTime'])
```

```python
dt['arrivalMonth'] = m[len(dt):]
dt['arrivalDate'] = d[len(dt):]
dt['arrivalHour'] = h[len(dt):]
dt['arrivalMinute'] = mm[len(dt):]
dt['arrivalDay'] = w[len(dt):]
```
